```python
import matplotlib.pyplot as plt
import tensorflow as tf
from PIL import Image
```

```python
model=tf.keras.models.load_model('./../Datasets/models/satellite_image_classification.h5')
```

```python

def predict(image):
    plt.imshow(Image.open(image))
    image=Image.open(image).resize((64,64))
    arr=tf.reshape(tf.divide(tf.constant(image),255),(-1,64, 64, 3))
    labels=('cloudy','desert','green_area','water')
    preds=model.predict(arr,verbose=0)
    preds=tf.round(preds*100,2).numpy()[0].tolist()
    print(dict(zip(labels,preds)))
    Max=max(preds)
    return f'{labels[preds.index(Max)]} {Max} %'


predict('ocean2.jpg')
```
